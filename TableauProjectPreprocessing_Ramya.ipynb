# Identify the outliers in the dataset

```python
def seperate_data_types(df):
    categorical = []
    continous = []
    for column in df.columns:
        if df[column].dtypes == object:
            categorical.append(column)
        else:
            continous.append(column)
    return categorical,continous
categorical,continous = seperate_data_types(df)


from tabulate import tabulate
table = [categorical,continous]
print(tabulate({'Categorical': categorical,'Continous':continous}, headers =['categorical','continous']))
```

```python
df_1 = df[['CLIENTNUM','Customer_Age','Dependent_count','Months_on_book','Total_Relationship_Count','Months_Inactive_12_mon','Contacts_Count_12_mon','Credit_Limit','Total_Revolving_Bal','Avg_Open_To_Buy','Total_Trans_Amt','Total_Trans_Ct','Avg_Utilization_Ratio']]
df_1
```

```python
for d in df_1:
    LO = []
    UO = []
    
    # Measuring Range
    data_series = pd.Series(df_1[d])
    Min = min(data_series)
    Max = max(data_series)
    Range = Max-Min
    
    Q1 = data_series.quantile(0.25)  # calculating 1st Quartile
    Q3 = data_series.quantile(0.75)  # calculating 3rd Quartile
    IQR = Q3-Q1                      # calculating Inter Quartile Range
    UF = Q3 + 1.5 * IQR              # calculating upper fence
    LF = Q1 - 1.5 * IQR              # calculating lower fence
    
    for i in df_1[d]:
        if (i<LF):
            LO.append(i)
        if (i>UF): 
            UO.append(i)

# Store the Min,Max,Range,Q1,Q2,IQR,LF,UF,Outliers into their respective columns.

    if (d == 'CLIENTNUM'):
        CLIENTNUM = ['CLIENTNUM',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Customer_Age'):
        Customer_Age = ['Customer_Age',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Dependent_count'):
        Dependent_count = ['Dependent_count',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Months_on_book'):
        Months_on_book = ['Months_on_book',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Total_Relationship_Count'):
        Total_Relationship_Count = ['Total_Relationship_Count',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Months_Inactive_12_mon'):
        Months_Inactive_12_mon = ['Months_Inactive_12_mon',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Contacts_Count_12_mon'):
        Contacts_Count_12_mon = ['Contacts_Count_12_mon',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Credit_Limit'):
        Credit_Limit = ['Credit_Limit',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Total_Revolving_Bal'):
        Total_Revolving_Bal = ['Total_Revolving_Bal',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Avg_Open_To_Buy'):
        Avg_Open_To_Buy = ['Avg_Open_To_Buy',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Total_Trans_Amt'):
        Total_Trans_Amt = ['Total_Trans_Amt',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    elif (d == 'Total_Trans_Ct'):
        Total_Trans_Ct = ['Total_Trans_Ct',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]
    else:
        Avg_Utilization_Ratio = ['Avg_Utilization_Ratio',Min,Max,Range,Q1,Q3,IQR,LF,UF,[LO,UO]]

col_names = ['Columns','Min','Max','Range','Q1','Q3','IQR','UF','LF','Outliers']
```

```python
data = pd.DataFrame([CLIENTNUM,Customer_Age,Dependent_count,Months_on_book,Total_Relationship_Count,
                     Months_Inactive_12_mon,Contacts_Count_12_mon,Credit_Limit,Total_Revolving_Bal,Avg_Open_To_Buy,
                     Total_Trans_Amt,Total_Trans_Ct,Avg_Utilization_Ratio], columns = col_names)
data
```

# Perform data imputation, using whichever method suits most logically for the given columns in the dataset

```python
def info_of_cat(col):
    print(f'Unique values in {col} are : {df[col].unique()}')
    print(f'mode of  {col} is : {df[col].mode()[0]}')
    print(f'total count of missing values in {col} is : {df[col].isnull().sum()}')
```

```python
def info_of_numerical(col):
    print(f'the mean of the {col} are : {df[col].mean()}')
    print(f'the median of the {col} are : {df[col].median()}')
    print(f'mode of  {col} is : {df[col].mode()[0]}')
    print(f'standard deviation of  {col} is : {df[col].std()}')
    print(f'total count of missing values in {col} is : {df[col].isnull().sum()}')
```

```python
info_of_cat('Card_Category')
```

```python
info_of_numerical('Credit_Limit')
```

```python
df['Card_Category'].fillna('Blue', inplace = True)
```

```python
df['Credit_Limit'].fillna(df['Credit_Limit'].mean(), inplace = True)
```
